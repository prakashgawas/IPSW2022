```python
import pandas as pd
import numpy as np
```

```python
month = ['January', 'February', 'March', 'April', 
         'May', 'June', 'July', 'August', 'September',
         'October', 'November', 'December']
days_in_month = [31, 28 , 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
df = pd.DataFrame(columns = [ 'Day','Month', 'Cycle'])
```

```python
cycle = {4,5,5}
for i in range(6,13):
    j = i % 12
    temp = pd.DataFrame(index = np.arange(len(df),len(df) + days_in_month[j]),columns = [ 'Day','Month', 'Cycle'])
    temp['Month'] = month[j]
    temp['Day'] = np.arange(1,days_in_month[j]+1)
    df = pd.concat([df,temp])
    
```
